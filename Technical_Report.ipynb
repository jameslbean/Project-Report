```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from scipy import stats
import seaborn as sns
df = pd.read_csv('titanic_train.csv')
```

```python

numerical_cols = ['Age']
df[numerical_cols].hist(bins=30, figsize=(12, 5))
plt.suptitle('Histograms of Age')
plt.show()
plt.savefig("figure.png")
```

```python
plt.figure(figsize=(8, 5))
df['Survived'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90)
plt.title('Survival Distribution')
plt.ylabel('')
plt.show()
```

```python
x = df['Pclass'].values
y = df['Fare'].values

slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)

line = slope * x + intercept

dot_size = 2
plt.scatter(x, y , color='blue')
plt.plot(x, line, color='red', label=f'Linear fit: y={slope:.2f}x+{intercept:.2f}')
plt.xticks(np.arange(min(x), max(x) + 1, 1))
plt.xlabel('Pclass')
plt.ylabel('Fare')
plt.title('Scatter Plot of Fare vs Age with Linear Regression Line')
plt.legend()
plt.grid(visible=True, linestyle='--', alpha=0.5)
plt.show()
```

```python
contingency_table = pd.crosstab(df['Survived'], df['Pclass'], margins=True)

chi2, p, dof, ex = stats.chi2_contingency(contingency_table)

print(f"Chi-square statistic: {chi2}")
print(f"P-value: {p}")
if p < 0.05:
    print("There is a significant association between Survival and Pclass.")
else:
    print("No significant association found.")
```
